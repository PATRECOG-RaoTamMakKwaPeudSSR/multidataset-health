```python
# 3 Datasets
df_statlog_chd_stroke_intersection = get_unscaled_combined("statlog-chd-stroke-intersection")
df_statlog_chd_stroke_union = get_unscaled_combined("statlog-chd-stroke-union")
df_statlog_chd_framingham_intersection = get_unscaled_combined("statlog-chd-framingham-intersection")
df_statlog_chd_framingham_union = get_unscaled_combined("statlog-chd-framingham-union")

# 2 Datasets
df_statlog_chd_intersection = get_unscaled_combined("statlog-chd-intersection")
df_statlog_chd_union = get_unscaled_combined("statlog-chd-union")
df_statlog_framingham_intersection = get_unscaled_combined("statlog-framingham-intersection")
df_statlog_framingham_union = get_unscaled_combined("statlog-framingham-union")
```

# Data Preprocessing

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

union_datasets = {
    "statlog-chd-stroke-union": df_statlog_chd_stroke_union,
    "statlog-chd-framingham-union": df_statlog_chd_framingham_union,
    "statlog-chd-union": df_statlog_chd_union,
    "statlog-framingham-union": df_statlog_framingham_union
}

intersection_datasets = {
    "statlog-chd-stroke-intersection": df_statlog_chd_stroke_intersection,
    "statlog-chd-framingham-intersection": df_statlog_chd_framingham_intersection,
    "statlog-chd-intersection": df_statlog_chd_intersection,
    "statlog-framingham-intersection": df_statlog_framingham_intersection
}

def process_union(df, name):
    df_processed = df.copy()

    numeric_cols = df_processed.select_dtypes(include=['number']).columns
    categorical_cols = df_processed.select_dtypes(exclude=['number']).columns

    if len(numeric_cols) > 0:
        num_imputer = SimpleImputer(strategy='median')
        df_processed[numeric_cols] = num_imputer.fit_transform(df_processed[numeric_cols])

    if len(categorical_cols) > 0:
        cat_imputer = SimpleImputer(strategy='most_frequent')
        df_processed[categorical_cols] = cat_imputer.fit_transform(df_processed[categorical_cols])

    scaler = MinMaxScaler()
    if len(numeric_cols) > 0:
        df_processed[numeric_cols] = scaler.fit_transform(df_processed[numeric_cols])

    set_preprocessed_combined(df_processed, scaler, name)
    print(f"✅ Processed Union Dataset: {name}")

def process_intersection(df, name):
    df_processed = df.copy()
    numeric_cols = df_processed.select_dtypes(include=['number']).columns

    scaler = MinMaxScaler()
    if len(numeric_cols) > 0:
        df_processed[numeric_cols] = scaler.fit_transform(df_processed[numeric_cols])

    set_preprocessed_combined(df_processed, scaler, name)
    print(f"✅ Processed Intersection Dataset: {name}")

print("--- Processing Union Datasets ---")
for name, df in union_datasets.items():
    process_union(df, name)

print("\n--- Processing Intersection Datasets ---")
for name, df in intersection_datasets.items():
    process_intersection(df, name)

print("\n🎉 All datasets processed and saved successfully!")
```
