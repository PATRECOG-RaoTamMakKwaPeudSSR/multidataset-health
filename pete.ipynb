```python
# 3 Datasets
df_statlog_framingham_stroke_intersection = get_unscaled_combined("statlog-framingham-stroke-intersection")
df_statlog_framingham_stroke_union = get_unscaled_combined("statlog-framingham-stroke-union")
df_statlog_heart_stroke_intersection = get_unscaled_combined("statlog-heart-stroke-intersection")
df_statlog_heart_stroke_union = get_unscaled_combined("statlog-heart-stroke-union")

# 2 Datasets
df_framingham_stroke_intersection = get_unscaled_combined("framingham-stroke-intersection")
df_framingham_stroke_union = get_unscaled_combined("framingham-stroke-union")
df_heart_stroke_intersection = get_unscaled_combined("heart-stroke-intersection")
df_heart_stroke_union = get_unscaled_combined("heart-stroke-union")
```

# Data Preprocessing

```python
def preprocess_intersection(df, name):
  numeric_cols = df.select_dtypes(include=['number']).columns
  df_new = df.copy()

  scaler = MinMaxScaler()
  df_new[numeric_cols] = scaler.fit_transform(df_new[numeric_cols])
  set_preprocessed_combined(df_new, name, scaler, f"{name}scaler")
  return df_new, scaler
```

```python
df_statlog_framingham_stroke_intersection_new, _ = preprocess_intersection(df_statlog_framingham_stroke_intersection, "statlog_framingham_stroke_intersection")
df_statlog_heart_stroke_intersection_new, scaler = preprocess_intersection(df_statlog_heart_stroke_intersection, "statlog_heart_stroke_intersection")
df_framingham_stroke_intersection_new, scaler = preprocess_intersection(df_framingham_stroke_intersection, "framingham_stroke_intersection")
df_heart_stroke_intersection_new, scaler = preprocess_intersection(df_heart_stroke_intersection, "heart_stroke_intersection")
```

```python
#"statlog-framingham-stroke-union"
df_statlog_framingham_stroke_union_new = df_statlog_framingham_stroke_union.copy()

drop_cols = ["cp_type", "fbs", "ecg", "angina", "oldpeak", "slope", "mv", "thal", "smoking_status", "hypertension_status", "bmi", "glucose_level"]
df_statlog_framingham_stroke_union_new = df_statlog_framingham_stroke_union_new.drop(columns=drop_cols)

mode_cols = ["bp_status", "stroke_status", "diabetes_status", "heart_disease_status", "marital_status", "residence_type", "work_type"]
mode_imputer = SimpleImputer(strategy="most_frequent")
df_statlog_framingham_stroke_union_new[mode_cols] = mode_imputer.fit_transform(df_statlog_framingham_stroke_union_new[mode_cols])

med_cols = ["edu", "num_cigs_per_day"]
med_imputer = SimpleImputer(strategy="median")
df_statlog_framingham_stroke_union_new[med_cols] = med_imputer.fit_transform(df_statlog_framingham_stroke_union_new[med_cols])

mean_cols = ["sbp", "chol", "hr", "dbp"]
mean_imputer = SimpleImputer(strategy="mean")
df_statlog_framingham_stroke_union_new[mean_cols] = med_imputer.fit_transform(df_statlog_framingham_stroke_union_new[mean_cols])

cols = df_statlog_framingham_stroke_union_new.select_dtypes(include=["number"]).columns
scaler = MinMaxScaler()
df_statlog_framingham_stroke_union_new[cols] = scaler.fit_transform(df_statlog_framingham_stroke_union_new[cols])

set_preprocessed_combined(df_statlog_framingham_stroke_union_new, "statlog_framingham_stroke_union", scaler, "statlog_framingham_stroke_union_scaler")
```

```python
#"statlog-heart-stroke-union"
df_statlog_heart_stroke_union_new = df_statlog_heart_stroke_union.copy()

drop_cols = ["cp_type", "sbp", "chol", "fbs", "ecg", "hr", "angina", "oldpeak", "slope", "mv", "thal"]
df_statlog_heart_stroke_union_new = df_statlog_heart_stroke_union_new.drop(columns=drop_cols)

mode_cols = ["hypertension_status", "heart_disease_status", "marital_status", "work_type", "residence_type", "smoking_status"]
mode_imputer = SimpleImputer(strategy="most_frequent")
df_statlog_heart_stroke_union_new[mode_cols] = mode_imputer.fit_transform(df_statlog_heart_stroke_union_new[mode_cols])

mean_cols = ["glucose_level", "bmi"]
mean_imputer = SimpleImputer(strategy="mean")
df_statlog_heart_stroke_union_new[mean_cols] = mean_imputer.fit_transform(df_statlog_heart_stroke_union_new[mean_cols])

scaler = MinMaxScaler()
cols = df_statlog_heart_stroke_union_new.select_dtypes(include=['number']).columns
df_statlog_heart_stroke_union_new[cols] = scaler.fit_transform(df_statlog_heart_stroke_union_new[cols])

set_preprocessed_combined(df_statlog_heart_stroke_union_new, "statlog_heart_stroke_union", scaler, "statlog_heart_stroke_union_scaler")
```
