# Data Preprocessing

```python
statlog_cat = ["sex", "cp_type" , "fbs" , "ecg" , "angina" ,"slope" ,"mv","thal" ,"target"]
chd_cat = ["famhist","target"]
framington_cat = ["sex","edu","smoking_status","bp_status","stroke_status" ,"hypertension_status" , "diabetes_status", "target"]
heart_cat = ["sex","cp_type" , "fbs","ecg","angina","slope","mv","thal","target"]
stroke_cat = ["sex", "hypertension_status" ,"heart_disease_status" , "martial_status" ,"work_type","resident_type","smoking_status","target"]
```

```python
#df_statlog_chd_heart_intersection.head()


list_cat = []
for i in statlog_cat:
    list_cat.append(i)

for i in chd_cat:
    if(i not in list_cat):
        list_cat.append(i)

for i in heart_cat:
    if(i not in list_cat):
        list_cat.append(i)
print(list_cat)

for col in df_statlog_chd_heart_intersection.columns:
    if(col in list_cat):
        df_statlog_chd_heart_intersection[col].fillna(df_statlog_chd_heart_intersection[col].mode()[0],inplace=True) 
    else:
        df_statlog_chd_heart_intersection[col].fillna(df_statlog_chd_heart_intersection[col].mean(),inplace=True) 

scaler = MinMaxScaler()
scaled_statlog_chd_heart_intersection = df_statlog_chd_heart_intersection.copy()

for col in df_statlog_chd_heart_intersection.columns:   
    if(col not in list_cat):
        scaled_statlog_chd_heart_intersection[col] = scaler.fit_transform(scaled_statlog_chd_heart_intersection[[col]])


print(scaled_statlog_chd_heart_intersection.to_string())
```
